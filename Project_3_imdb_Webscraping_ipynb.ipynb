```python
#import all the packages
import requests
import time
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# Get the HTML content from the web page and parsing the HTML content
url = 'https://www.imdb.com/search/title/?release_date=2021&sort=num_votes,desc&page='
req = requests.get(url)
soup = BeautifulSoup(req.text,'html')
```

```python
# to separate meta score and None
Final_metascore=[]
for i in meta_tag:
  m=i.find('span',attrs={'class':'metascore'})
  if m is not None:
    Final_metascore.append(m.text.strip())
  else:
      Final_metascore.append('None')


print(Final_metascore)  
```

```python
# to separate certificates and None
Final_cere=[]
for i in soup.findAll('div',attrs={'class':'lister-item-content'}):
  m=i.find('span',attrs={'class':'certificate'})
  if m is not None:
    Final_cere.append(m.text)
  else:
      Final_cere.append('None')


len(Final_cere)  
#Final_cere
```
